```python
df=pd.read_excel("C:\\Users\\ANKIT\\Desktop\\Excel_DATA.xlsx",sheet_name="Data",skiprows=1)
df
```

```python
df['firstname']=='papa'          # filter in the form of true and false
```

```python
# Filter
filt=(df['firstname']=='papa')             # filtering data
df[filt]
```

```python
filt=((df['firstname']=='papa') & (df['Age']==12))
# & = And  (in and codition : if both condition are true then it gives output )
df[filt]
```

```python
filt=((df['firstname']=='papa') | (df['Age']==12))
df[filt]
# | = Or ( in or contion it give output if one condition is true and other is false)
```

```python
Agessss=[12,13,1,2,12,13,12,12,15]   # assigning own age

filt= df['Age'].isin(Agessss)         # getting data with assign age        
df[filt]             # .isin is used for checking assign data in data frame
```

```python
filt=df['lastname'].str.contains('kumar')            # str.contains is used for filtering specific name data
df[filt]
```

```python
df.columns=[x.upper() for x in df.columns]         # Getting all the columns in Upper case
df
```

```python
df.columns=[x.lower() for x in df.columns]         # Getting all the columns in lower case
df
```

```python
df.columns=df.columns.str.replace('age','Age')         # Modifying the columns name 
df                                                     # .str.replace is used for replace the column name


##### Modifying is permanent ( changes are done in original df)
```

```python
## Rename the the multiple columns
df.rename(columns={'salary':'Salary',            
                  'Age':'age'})

## Changes are done only in a block (temporary changes), not in df
```
